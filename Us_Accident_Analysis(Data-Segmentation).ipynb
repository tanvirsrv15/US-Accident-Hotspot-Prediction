```python
missing_values = Us_Accident.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
# Calculate the state counts
state_counts = Us_Accident['State'].value_counts()

state_percentage = (state_counts / state_counts.sum()) * 100
print(state_percentage)
```

```python
state_files = {
    'FL': 'Us_Accident_FL.csv',
    'CA': 'Us_Accident_CA.csv',
    'TX': 'Us_Accident_TX.csv',
    'SC': 'Us_Accident_SC.csv',
    'NY': 'Us_Accident_NY.csv',
    'NC': 'Us_Accident_NC.csv'
}

for state, filename in state_files.items():
    filtered_df = Us_Accident[Us_Accident['State'] == state]
    filtered_df.to_csv(filename, index=False)
```
